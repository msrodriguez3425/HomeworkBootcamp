```python
#Import Dependencies
import pandas as pd
import numpy as np
```

```python
#Assign Filepaths to read CSVs
schoolsPath = "Resources/schools_complete.csv"
studentsPath = "Resources/students_complete.csv"
```

```python
#Read the schools CSV
schools_df = pd.read_csv(schoolsPath)
#schools_df.head()
```

```python
#Read the Students CSV
students_df = pd.read_csv(studentsPath)
#students_df.head()
```

```python
#Merge students and schools data frames into one single dataframe
schools_students_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
schools_students_df.head()
```

```python
#1. District Summary
```

```python
#Total Schools
totalSchools = len(schools_students_df["school_name"].unique())
```

```python
#Total Students
totalStudents = schools_students_df["student_name"].count()
```

```python
#Total Budget
totalBudget = (schools_students_df["budget"].unique()).sum() 
```

```python
#Average Math Score
averageMath = schools_students_df["math_score"].mean()
```

```python
#Average Reading Score
averageReading = schools_students_df["reading_score"].mean()
```

```python
#% Passing Math
num_pass_math = len(schools_students_df.loc[schools_students_df["math_score"]>=70])
percent_pass_math = (num_pass_math/totalStudents)*100
```

```python
#% Passing Reading
num_pass_reading = len(schools_students_df.loc[schools_students_df["reading_score"]>=70])
percent_pass_reading = (num_pass_reading/totalStudents)*100
```

```python
#Overall Passing Rate
overallPass = (averageMath+averageReading)/2
```

```python
district_df = pd.DataFrame([{"Total Schools":totalSchools,
    "Total Students":totalStudents,
    "Total Budget":totalBudget,
    "Average Math Score":averageMath,
    "Average Reading Score":averageReading,
    "% Passing Math":percent_pass_math,
    "% Passing Reading":percent_pass_reading,
    "% Overall Passing Rate":overallPass}])

district_df
```

```python
#2. School Summary
```

```python
#Group the data by school_name
grouped_schools_df = schools_students_df.groupby(['school_name'])
```

```python
#Students in each school
students_per_school = grouped_schools_df["student_name"].count()
```

```python
#Average math score in each school
school_av_math =grouped_schools_df["math_score"].mean()
```

```python
#Average reading score in each school
school_av_reading = grouped_schools_df["reading_score"].mean()
```

```python
#Budget of each school
school_budget = grouped_schools_df["budget"].sum()/students_per_school
```

```python
#Budget of school divided by number of students in school
budget_per_student = school_budget/students_per_school
```

```python
#School types
school_types = grouped_schools_df["type"].first()
```

```python
#To get percentage passing rates:
```

```python
#Adding new columns to original merged data frame, that identifies (using bins) 
#students who passed math and/or reading or neither.
bins = [0,69,100]
group_names=[0,1]

schools_students_df["passed_Math"] = pd.cut(schools_students_df["math_score"],bins,labels=group_names)
schools_students_df["passed_Reading"] = pd.cut(schools_students_df["reading_score"],bins,labels=group_names)
```

```python
schools_students_df.head()
```

```python
#Converting passed_Math and passed_Reading columns to numeric so they can be summed.
schools_students_df["passed_Math"] = pd.to_numeric(schools_students_df["passed_Math"])
schools_students_df["passed_Reading"] = pd.to_numeric(schools_students_df["passed_Reading"])
```

```python
#Grouping again
grouped_2 = schools_students_df.groupby(["school_name"])
grouped_2.count().head()
```

```python
#Summing 1s and 0s per school
passed_math = grouped_2["passed_Math"].sum()
passed_reading = grouped_2["passed_Reading"].sum()
```

```python
#Percent that passed math in each school
percent_passed_math = (passed_math/students_per_school)*100
```

```python
#Percent that passed reading in each school
percent_passed_reading = (passed_reading/students_per_school)*100
```

```python
#Overall pass rate in each school
overall_average = (percent_passed_reading + percent_passed_math)/2
```

```python
#School Summary Table, making a data frame out of all the variables calculated
school_summary_table = pd.DataFrame({"School Type":school_types,
                                    "Total Students": students_per_school,
                                    "Math Average": school_av_math,
                                    "Reading Average": school_av_reading,
                                    "School Budget": school_budget,
                                    "Budget Per Student":budget_per_student,
                                    "Math Pass Rate": percent_passed_math,
                                    "Reading Pass Rate": percent_passed_reading,
                                    "Overall Pass Rate":overall_average
                                   })
school_summary_table.head()
```

```python
#Top Perfoming Schools (By Passing Rate)
school_ranking_df = school_summary_table.sort_values("Overall Pass Rate", ascending=False)
school_ranking_df.head()
```

```python
#Bottom Performing Schools (By Passing Rate)
school_ranking_df = school_summary_table.sort_values("Overall Pass Rate")
school_ranking_df.head()
```

```python
#Scores by school Spending
```

```python
#Establish bins and group names
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#Gather relevant data
by_ranges_table = school_summary_table.reset_index()
by_ranges_table = by_ranges_table[["Budget Per Student","Math Average","Reading Average","Math Pass Rate",
                                  "Reading Pass Rate","Overall Pass Rate"]]
by_ranges_table.head()
```

```python
#Gather data in bins using cut function
by_ranges_table["Spending Ranges (Per Student)"] = pd.cut(by_ranges_table["Budget Per Student"],
                                                          spending_bins,labels=group_names)
by_ranges_table.head()
```

```python
#Group by the spending ranges
av_ranges_table = by_ranges_table.groupby(["Spending Ranges (Per Student)"]).mean()
```

```python
#Final Scores by School Spending Table
final_ranges_table = av_ranges_table.drop("Budget Per Student",axis = 1)
final_ranges_table.head()
```

```python
#Scores by School Size
```

```python
# Establish bins and group names
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#Gathering relevant data
by_size = school_summary_table.reset_index()
by_size = by_size[["Total Students","Math Average","Reading Average","Math Pass Rate","Reading Pass Rate","Overall Pass Rate"]]
```

```python
#Gathering data into bins using cut function
by_size["School Size"] = pd.cut(by_size["Total Students"],size_bins,labels = group_names)
by_size.head()
```

```python
#Grouping data by School size
grouped_by_size = by_size.groupby(["School Size"]).mean()
grouped_by_size.head()
```

```python
#Final Scores by School Size Table
final_grouped_by_size = grouped_by_size.drop("Total Students",axis = 1)
final_grouped_by_size
```
